```python
import cv2
print(cv2.__version__)
```

```python
#downloading file from url
import urllib.request

trumptredeu_url = "https://upload.wikimedia.org/wikipedia/commons/e/ee/Donald_Trump_and_Justin_Trudeau_in_the_Oval_Office_-_2017.jpg"
trumptredeu_filename = r"trumptredeu.jpg"

urllib.request.urlretrieve(trumptredeu_url,trumptredeu_filename)
```

```python
import os
os.listdir(os.curdir)
```

```python
print(trumptredeu_filename in os.listdir(os.curdir))
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

trumptredeu = cv2.imread(trumptredeu_filename)
plt.imshow(trumptredeu)
```

```python
img_corrected = cv2.cvtColor(trumptredeu,cv2.COLOR_BGR2RGB)
plt.imshow(img_corrected)
```

```python
#removethe axis around the image
plt.axis("off")
plt.imshow(img_corrected)
```

```python
from pylab import rcParams
rcParams["figure.figsize"] = 10,12
plt.axis("off")
plt.imshow(img_corrected)
```

```python
gray_trumptrudeu = cv2.cvtColor(trumptredeu, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_trumptrudeu,cmap='gray')
plt.axis("off")
plt.title("Gray scale image")
```

```python
rcParams["figure.figsize"] = 10,12
edge = cv2.Canny(img_corrected, threshold1 = 100, threshold2 = 200)

plt.imshow(edge,cmap="gray")
plt.title("Canny edge detection")
plt.xticks([]),plt.yticks([])
```

```python
rcParams["figure.figsize"] = 10,12
edge = cv2.Canny(img_corrected, threshold1 = 100, threshold2 = 500)

plt.imshow(edge,cmap="gray")
plt.title("Canny edge detection")
plt.xticks([]),plt.yticks([])
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

rcParams['figure.figsize'] = 8,10
plt.hist(gray_trumptrudeu.ravel(),256,[0,256])
plt.title('Raveled histogram of Trump and Trudeu')
plt.show()
```

```python
rcParams['figure.figsize'] = 8,6
color = ['r','g','b']
for i,col in enumerate(color):
    histr = cv2.calcHist([trumptredeu],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.show()
```
